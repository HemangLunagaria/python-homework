```python
import csv
from pathlib import Path

menu_data_path = Path('./Resources/menu_data.csv')
sales_data_path = Path('./Resources/sales_data.csv')

report_path = Path("./Results/report.txt")

menu_data = []
sales_data = []
report = {}

with open(menu_data_path, 'r') as file:
    rows = csv.reader(file, delimiter = ",")
    next(rows, None)
    menu_data = list(rows)
    file.close()

with open(sales_data_path, 'r') as file:
    rows = csv.reader(file, delimiter = ",")
    next(rows, None)
    sales_data = list(rows)
    file.close()
    
for menu in menu_data:
    item = menu[0]
    price = float(menu[3])
    cost = float(menu[4])
    profit = price - cost
    not_found_ctr = 0

    for sale in sales_data:
        sale_menu_item = sale[4]
        quantity = int(sale[3])

        if sale_menu_item not in report.keys():
            report[sale_menu_item] = {
                "01-count" : 0,
                "02-revenue" : 0,
                "03-cogs" : 0,
                "04-profit" : 0
            }

        if item == sale_menu_item:
            not_found_ctr = 0
            report[sale_menu_item]["01-count"] += quantity
            report[sale_menu_item]["02-revenue"] += price * quantity
            report[sale_menu_item]["03-cogs"] += cost * quantity
            report[sale_menu_item]["04-profit"] += profit * quantity
        else:
            not_found_ctr += 1
    
    if not_found_ctr >= (len(sales_data) -1) :
        print(f"{item} not in Sales data! NO MATCH\n")
        

with open(report_path, 'w') as file:
    for key in report.keys():
        result = key + " " + str(report[key]) + "\n"
        file.write(result)
        print(result)


    
    
```
